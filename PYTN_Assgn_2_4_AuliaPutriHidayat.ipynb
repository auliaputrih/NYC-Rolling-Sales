## Data Preprocessing

##### Langkah untuk mengubah maupun memasukkan data ke dalam sebuah data set. Meminimalkan data yang salah atau tidak relevan yang dapat berdampak pada statistik data. 

```python
#Menghapus kolom data yang tidak diperlukan
df = df.drop(["Unnamed: 0",
              "NEIGHBORHOOD", 
              "BUILDING CLASS CATEGORY",
              "TAX CLASS AT PRESENT",
              "EASE-MENT",
              "BUILDING CLASS AT PRESENT",
              "ADDRESS", 
              "APARTMENT NUMBER", 
              "ZIP CODE"], axis=1)

df
```

```python
#Melakukan convert tipe data pada kolom
#Convert object -> int64
convertInt=["LAND SQUARE FEET", "GROSS SQUARE FEET", "SALE PRICE"]
for i in convertInt:
    df[i] = pd.to_numeric(df[i], errors="coerce").fillna(0).astype("int64")

#Convert object -> datetime
convertDate=["SALE DATE"]
for i in convertDate:
    df[i] = pd.to_datetime(df[i], errors="coerce")

df.dtypes
```

```python
#Melakukan perubahan pada kolom borough
df["BOROUGH"] = df.BOROUGH.replace({1: "Manhattan", 2: "Bronx", 3: "Brooklyn", 4: "Queens", 5: "Staten Island"})
df
```

```python
#Mengganti nilai pada kolom year built dengan nilai median sebagai langkah menghilangkan outlier
mean_yearbuilt = df["YEAR BUILT"].median()
mean_yearbuilt
```

```python
df["YEAR BUILT"]=df["YEAR BUILT"].mask(df["YEAR BUILT"] == 0,1940)
df
```

```python
#Mengganti nilai pada kolom sale price dengan nilai median sebagai langkah menghilangkan outlier
mean_saleprice = df["SALE PRICE"].median()
mean_saleprice
```

```python
df["SALE PRICE"]=df["SALE PRICE"].mask(df["SALE PRICE"] == 0,415000)
df
```

## Exploratory Data Analysis

### Measure of Central Tendency

### Mean (Rata-rata)

##### Bagaimana nilai mean pada kolom land square feet, gross square feet, dan sale price di setiap wilayah (borough)?

```python
#Menghitung nilai mean LAND SQUARE FEET dikelompokkan berdasarkan data di kolom BOROUGH
meanLandSquare = df[["BOROUGH","LAND SQUARE FEET"]].groupby("BOROUGH").mean()
meanLandSquare
```

```python
#Menghitung nilai mean GROSS SQUARE FEET dikelompokkan berdasarkan data di kolom BOROUGH
meanGrossSquare = df[["BOROUGH","GROSS SQUARE FEET"]].groupby("BOROUGH").mean()
meanGrossSquare
```

```python
#Menghitung nilai mean SALE PRICE dikelompokkan berdasarkan data di kolom BOROUGH
meanPrice = df[["BOROUGH","SALE PRICE"]].groupby("BOROUGH").mean().sort_values(by="SALE PRICE", ascending=True)
meanPrice
```

### Median

##### Berapa nilai median pada kolom residential units, total units, sale price di setiap wilayah (borough)?

```python
#Menghitung nilai median RESIDENTIAL UNITS pada setiap data di kolom BOROUGH
medianResidential = df[["BOROUGH","RESIDENTIAL UNITS"]].groupby("BOROUGH").median().sort_values(by="RESIDENTIAL UNITS", ascending=True)
medianResidential
```

```python
#Menghitung nilai median TOTAL UNITS pada setiap data di kolom BOROUGH
medianTotalUnit = df[["BOROUGH","TOTAL UNITS"]].groupby("BOROUGH").median().sort_values(by="TOTAL UNITS", ascending=True)
medianTotalUnit
```

```python
#Menghitung nilai median SALE PRICE pada setiap data di kolom BOROUGH
medianSalePrice = df[["BOROUGH","SALE PRICE"]].groupby("BOROUGH").median().sort_values(by="SALE PRICE", ascending=True)
medianSalePrice
```

### Modus

##### Berapa nilai modus pada kolom residential units, commercial units, total units, land square feet, gross square feet, dan sale price di setiap wilayah (borough)?

### Hypothesis Testing

##### Apakah ada perbedaan unit yang terjual antara properti yang dibangun tahun 1900-2000 dengan tahun 2001?

```python
data_nyc = df.copy()

#Mendapatkan data tahun dibangun antara tahun 1900-2000 dan 2001
sales1900_2000 = data_nyc[(data_nyc["YEAR BUILT"] >= 1900) & (data_nyc["YEAR BUILT"] <= 2000)]
sales2001 = data_nyc[data_nyc["YEAR BUILT"] == 2001]

#Menghitung rata-rata total unit yang terjual
mean1900_2000 = sales1900_2000["TOTAL UNITS"].mean()
mean2001 = sales2001["TOTAL UNITS"].mean()

print(f"Nilai mean unit yang terjual tahun 1900 dan 2000: {mean1900_2000:.2f}")
print(f"Nilai mean unit yang terjual tahun 2001: {mean2001:.2f}")
```

Kesimpulan dari hipothesis di atas ialah:
- Benar adanya bahwa terdapat perbedaan unit yang terjual antara properti tahun 1900 hingga 2000, dan properti tahun 2001. Diperoleh hasil bahwa unit properti yang terjual lebih banyak pada tahun 1900-2000.
